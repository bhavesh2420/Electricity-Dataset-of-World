```python
a=[]
b=[]
c=[]
d=[]
e=[]
f=[]
g=[]
h=[]
i=[]
j=[]
first_50 = [a,b,c,d,e,f,g,h,i,j]
inc = 0
for i in data['Entity'].unique()[0:50]:
    if len(first_50[inc])<5:
        first_50[inc].append(i)
    else:
        inc = inc+1
        first_50[inc].append(i)
```

```python
plt.style.use('fivethirtyeight')
fig, ax= plt.subplots(10,5,figsize=(30,50),facecolor='white')
for i in range(0,10):
    data1 = first_50[i]
    for j in range(0,5):
        ax[i,j].plot(data['Year'][data['Entity']==data1[j]],data['Per capita electricity (kWh)'][data['Entity']==data1[j]])
        ax[i,j].set_title(data1[j])
```

```python
x = india_electricity_consumption.drop(['Per capita electricity (kWh)', 'Entity'], axis = 1)
y = india_electricity_consumption['Per capita electricity (kWh)']
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
model_india=LinearRegression()
model_india.fit(x,y)
```

```python
print("Training Accuracy :", model_india.score(X_train, y_train))
print("Testing Accuracy :", model_india.score(X_test, y_test))
```

```python
print("Coefficient: ",model_india.coef_)
print("intercept: ",model_india.intercept_)
pre = model_india.predict(x)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x,y,label='Acutal value')
plt.plot(x,pre,color='g',label='Predicted value')
plt.legend()
plt.title("India",color='r')
plt.xlabel("Years",color='r')
plt.ylabel("per capital electricity consumption(kWh)",color='r')
plt.tight_layout()
plt.show()
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
years=[2021,2022,2023,2024,2025]
for i in years:
    print(model_india.predict([[i]]))
```

```python
from sklearn.metrics import mean_squared_error
```

```python
print("By function: ",mean_squared_error(y,model_india.predict(x)))
```
